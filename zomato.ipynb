```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly as px
```

```python
df=pd.read_csv("Zomato.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
def handleRate(value):
    value=str(value).split("/")
    value=value[0]
    return float(value)
    
df['rate']=df['rate'].apply(handleRate)
df.head()
```

```python
df.describe()
```

```python
df.info()
```

# types of restorent

```python
sns.countplot(x='listed_in(type)', data=df)
plt.xticks(fontsize=7)
plt.xlabel("Types of restorents")
plt.show()
```

# conclusion- Dining type restorent are mosorty numbers 

```python
group =df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes' : group})
plt.plot(result,marker='o')
plt.xlabel("restorent types")
plt.ylabel("votes")
plt.xticks(fontsize=7)
plt.show()
```

```python
plt.hist(df["rate"], bins=10 )
plt.xticks(fontsize=6)
plt.show()
```
